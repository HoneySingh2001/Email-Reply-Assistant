```python
!pip install transformers sentencepiece --quiet

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load free model from Hugging Face
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

nlp_pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

def generate_reply(email_text, tone="polite"):
    prompt = f"""
You are an assistant that writes {tone} and professional email replies.

Email received:
\"{email_text}\"

Write a reply.
"""
    result = nlp_pipeline(prompt, max_new_tokens=200)[0]['generated_text']
    return result

# Test it
email = "Hey, just wanted to check if you're available for a quick call tomorrow morning regarding the design review?"
print(generate_reply(email, tone="friendly"))
```
